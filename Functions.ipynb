1. What is the difference between a function and a method in Python?
*  A function is a standalone block of code that performs a specific task, while a method is a function that is associated with an object and can access its data.

2. Explain the concept of function arguments and parameters in Python.
*  In Python, parameters are the variables defined in a function's signature that accept values when the function is called, while arguments are the actual values passed to the function when it is invoked.

Example, in the function definition def add(a, b):, a and b are parameters. When you call add(3, 5), the values 3 and 5 are the arguments provided to the function.

6. Explain the concept of generators in Python and how they are defined.
*  Generators in Python are a special type of iterable that allow you to create iterators in a memory-efficient way.They are defined using functions but use the yield statement instead of return. When a generator function is called, it does not execute the function body immediately; instead, it returns a generator object that can be iterated over.
example:

def countdown(n):
    while n > 0:
        yield n  
        n -= 1   
for number in countdown(5):
    print(number)
Output: 5, 4, 3, 2, 1

7. What are the advantages of using generators over regular functions?
*  Unlike regualar functions, Generators save memory by producing one value at a time instead of storing everything at once. They make it easier to work with large or endless lists of data because they only create values when needed.

8.  What is a lambda function in Python and when is it typically used?
*   A lambda function in Python is a small, anonymous function defined with the lambda keyword, allowing for quick, single-expression functions. They are typically used for short operations, especially as arguments to higher-order functions like map(), filter(), and sorted(), where a simple function is needed temporarily.

example : add = lambda x, y: x + y
print(add(3, 5))
output: 8

9. Explain the purpose and usage of the `map()` function in Python.
*  Purpose of map is to apply a function to all items in an iterable, producing a new iterable with the results. Usage is : The map() function takes two arguments: the function to apply and the iterable. You can also pass multiple iterables if the function takes more than one argument.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
*  The map() function applies a specified function to each item in an iterable, transforming the data, while filter() selects items based on a condition, returning only those for which the function is True. The reduce() function, on the other hand, applies a function to the items of an iterable, reducing it to a single value, such as a sum or product.

```python
#Question 1
def s(numbers):
    return sum(num for num in numbers if num % 2 == 0)
numbers = [47, 11, 42, 13, 8, 20]
result = s(numbers)
print(result)
```

```python
#Question 2
def rs(s):
    return s[::-1]

input_string = "Hello, World!"
rs = rs(input_string)
print(rs)
```

```python
#Question 3
def sq(numbers):
    return [num ** 2 for num in numbers]

input_list = [1, 2, 3, 4, 5]
squared = sq(input_list)
print(squared)
```

```python
#Question 4
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(number, "is a prime number.")
```

```python
#Question 6
def powers(x):
    for i in range(x + 1):
        yield 2 ** i

# Example usage:
for power in powers(5):
    print(power)
```

```python
#Question 7
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()


file_path = 'example.txt'
for line in read_file(file_path):
    print(line)
```

```python
#Question 9
def convert(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 20, 37, 100]

fahrenheit_temperatures = list(map(convert, celsius_temperatures))

print(fahrenheit_temperatures)
```

```python
#Question 10
def vow(char):
    return char.lower() not in 'aeiou'

input_string = "Hello, World!"

filtered_string = ''.join(filter(vow, input_string))

print(filtered_string)
```
